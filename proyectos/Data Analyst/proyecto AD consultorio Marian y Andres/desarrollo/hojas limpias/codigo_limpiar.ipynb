```python
# codigo para todas las hojas
# hoja de detalle
import pandas as pd
import numpy as np
import os

ruta = "C:/Users/chris/OneDrive/Desktop/proyecto analisis de datos/desarrollo/Hoja_madre_detalle.xlsx"
data = pd.read_excel(ruta)
data1 = data
# === LIMPIEZA ===
def limpiar_hoja(df):
    # Eliminar columnas totalmente vacías
    df = df.dropna(axis=1, how="all")

    # Eliminar filas duplicadas
    df = df.drop_duplicates()

    # Eliminar filas con valores faltantes
    df = df.dropna()

    # Quitar outliers numéricos (IQR)
    for col in df.select_dtypes(include="number"):
        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)
        iqr = q3 - q1
        mask = df[col].between(q1 - 1.5 * iqr, q3 + 1.5 * iqr)
        df = df[mask]

    # Normalizar textos: eliminar espacios y minúsculas
    for col in df.select_dtypes(include="object"):
        df[col] = df[col].astype(str).str.strip().str.lower()

    return df

# === PROCESAMIENTO GENERAL ===


# === EJECUCIÓN ===
limpiar_hoja(data)
data1.shape
data.shape
print (data1)
rutag = "C:/Users/chris/OneDrive/Desktop/proyecto analisis de datos/desarrollo/hojas de limpias/" 
            "/Hoja_madre_detalle_limpia.xlsx"
data.to_excel(rutag, index=False)
```
